```python
import pandas as pd
```

```python
data=pd.read_csv("https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv")
```

```python
data
```

```python
def convert(cols):
    if cols[0]=='male':
        return 1
    else:
        return 0
    
data["Sex"]=data[['sex']].apply(convert,axis=1)
```

```python
data
```

```python
total_length=len(data['Sex'])-1
```

```python
plt.pie(list(data.groupby('Sex').size()),labels=['male','female'])
```

```python
mapping={1:'blue',0:'red'}
plt.scatter(data['age'],data['fare'],c=data['Sex'].apply(lambda x: mapping[x]))
```
